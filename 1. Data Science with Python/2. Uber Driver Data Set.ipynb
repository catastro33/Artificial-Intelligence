### Groupby
We are going to groupby the start column and we are going to view how many rides were taken in total from that particular place.

A rule of thumb while using a groupby object is that, the groupby method should always be followed by an aggregation method. 
In pandas, aggregation methods are used to summarize and compute statistics on data. Some of the common aggregation methods in pandas include:
- sum()
- mean()
- median()
- min()
- max()
- count()
- std()
- var()
- summarize()
- agg()
- value_counts()
- nunique()

```python
df_group = uber_df.groupby("START*")
```

```python
df_group.count()
```

```python
uber_df.groupby("START*")["MILES*"].agg("mean").head(n=20)
```

```python
uber_df.groupby("START*")["MILES*"].agg(["mean", "max"]).head(n=20)
```

Outliers - Outliers in a dataset are data points that significantly differ from the majority of the data points. In other words, they are observations that appear to be unusually distant from the central tendency of the data, such as the mean or median.

### Outlier Detection & Removal

```python
sample_data = pd.DataFrame([[150], [160], [170], [180], [185], [140], [130], [120], [190], [300]], columns = ["Height in CM"])
```

```python
sample_data
```

```python
sample_data.reset_index(level=0, inplace=True)
```

We are adding an index column because we wanted to visualize the scatter plot and make it more readable. In the first sample data output we can see that the table does not have a dedicated index column. The other column is just the label.

```python
sample_data
```

```python
import seaborn as sns
sns.scatterplot(x="index", y="Height in CM", data=sample_data)
```

```python
# This is the plot that is used whenever we want to detect outliers.

sns.boxplot(sample_data["Height in CM"])
```

```python
def outlier_detection(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn , [25,75]) # Q1- first quartile value, Q3- third quartile value
    IQR = Q3 - Q1 # IQR - Inter quartile range 
    lower_range = Q1 - (1.5 * IQR) # min permissible limit
    upper_range = Q3 + (1.5 * IQR) # max permissible limit
    return lower_range,upper_range
```

Q1 - (1.5 * IQR) - Help to include the values within the range 80 to 100

Q3 + (1.5 * IQR) - Help to include the values within the range 150 to 160

```python
lower, upper = outlier_detection(sample_data["Height in CM"])
```

```python
print("The lower range is", lower)
print("The upper range is", upper)
```

```python
sample_data[ (sample_data["Height in CM"] > upper) | (sample_data["Height in CM"] < lower) ]
```

```python
sample_data.drop(sample_data[ (sample_data["Height in CM"] > upper) | (sample_data["Height in CM"] < lower) ].index , inplace=True)
```
