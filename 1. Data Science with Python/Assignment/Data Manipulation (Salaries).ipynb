**How much salary cost has increased from 2011 to 2014**

```python
pay_2011 = df_salaries.query("Year == 2011")['TotalPay'].sum()
pay_2011
```

```python
pay_2014 = df_salaries.query("Year == 2014")['TotalPay'].sum()
pay_2014
```

```python
subtract = pay_2014 - pay_2011
subtract
```
