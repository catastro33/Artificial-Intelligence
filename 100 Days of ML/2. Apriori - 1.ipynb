```python
df = [["Milk", "Cake", "Bread", "Eggs", "Tea", "Sugar"],
      ["Knife", "Cake", "Bread", "Eggs", "Tea", "Sugar"],
      ["Milk", "Biscuits", "Eggs", "Tea"],
      ["Milk", "Salt", "Coco", "Eggs", "Sugar"],
      ["Salt", "Cake", "Bread", "Eggs", "Cream", "Sugar"]
     ]
```

```python
df
```

```python
type(df)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
te_array = te.fit_transform(df)
```

```python
te_array
```

```python
te.columns_
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
freq_itemsets = apriori(df, use_colnames=True)
```

```python
freq_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules = association_rules(freq_itemsets)
```

```python
rules
```

```python
rules = rules[["antecedents", "consequents", "support", "confidence"]]
```
