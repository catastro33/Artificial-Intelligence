## **Missing Indicator**

```python
import numpy as np
import pandas as pd
```

```python
titanic_df = pd.read_csv('D:/Academics/AI/Data Sets/Titanic.csv', usecols=['Age', 'Fare', 'Survived'])
```

```python
titanic_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = titanic_df.drop(columns=['Survived'])
y = titanic_df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.impute import SimpleImputer, MissingIndicator
```

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()

lr_model.fit(X_train_trf, y_train)

y_pred = lr_model.predict(X_test_trf)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
mi_model = MissingIndicator()
```

```python
mi_model.fit(X_train)
```

```python
mi_model.features_
```

```python
X_train_missing = mi_model.transform(X_train)
```

```python
X_train_missing
```

```python
X_test_missing = mi_model.transform(X_test)
```

```python
X_test_missing
```

```python
X_train['Age_NA'] = X_train_missing
```

```python
X_train.head()
```

```python
X_test['Age_NA'] = X_test_missing
```

```python
X_train
```

```python
X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogR = LogisticRegression()
```

```python
LogR.fit(X_train_trf2, y_train)
y_pred = LogR.predict(X_test_trf2)
```

```python
accuracy_score(y_test, y_pred)
```

```python
SimI = SimpleImputer(add_indicator=True)
```

```python
X_train = SimI.fit_transform(X_train)
```

```python
X_test = SimI.transform(X_test)
```

```python
LogR.fit(X_train_trf2,y_train)
y_pred = LogR.predict(X_test_trf2)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
